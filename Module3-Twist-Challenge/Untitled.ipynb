```python
#acled_data['country'].value_counts(normalize=True)
africa = acled_data[acled_data['region'].str.contains("Africa")]
africa['country'].unique()
```

```python
africa.groupby('country')['country'].count().reset_index(name='conflict_counts').sort_values(['conflict_counts'],ascending=False)
```

```python
africa.groupby('country')['fatalities'].sum().reset_index(name='fatalities_counts').sort_values(['fatalities_counts'],ascending=False)

```

```python
somalia = africa[africa['country']=='Somalia']
somalia['date'] = pd.to_datetime(somalia['event_date'])
somalia['year_mon'] = somalia['date'].dt.to_period('M')
somalia.head()
```

```python
plt.figure(figsize=(12,8))
somalia.groupby('year_mon')['fatalities'].sum().plot()
```
